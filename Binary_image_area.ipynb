```
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```
urls = ["https://www.armytimes.com/resizer/qA-v-SjHmT4r4tcKz-f4g3jFSuo=/1200x0/filters:quality(100)/arc-anglerfish-arc2-prod-mco.s3.amazonaws.com/public/Z3M4XZEHXJFGBCTZM2WNY5JAJQ.jpg"]
for url in urls:
  image = io.imread(url)
  gray1 = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
  print(gray1.shape)
  cv2_imshow(gray1)
```

```
ret, thresh1 = cv.threshold(gray1,113,255,cv.THRESH_BINARY)
print(thresh1.shape)
cv2_imshow(thresh1)
```
