```
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```
urls = ["https://www.israel21c.org/wp-content/uploads/2018/08/shutterstock_leukemia-1168x657.jpg"]
for url in urls:
  image = io.imread(url)
  gray1 = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
  print(gray1.shape)
  cv2_imshow(gray1)
```
