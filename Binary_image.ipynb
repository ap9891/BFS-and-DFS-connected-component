```
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```
urls = ["https://thumbor.forbes.com/thumbor/960x0/https%3A%2F%2Fblogs-images.forbes.com%2Fstartswithabang%2Ffiles%2F2020%2F04%2Fopo9601c-1200x947.jpg"]
for url in urls:
  image = io.imread(url)
  gray1 = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
  print(gray1.shape)
  cv2_imshow(gray1)
```

```
ret, thresh1 = cv.threshold(gray1,160,255,cv.THRESH_BINARY)
print(thresh1.shape)
cv2_imshow(thresh1)
```
